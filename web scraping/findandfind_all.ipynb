```python
from bs4 import BeautifulSoup
import requests
```

```python
path="https://www.scrapethissite.com/pages/forms/"
```

```python
page=requests.get(path)
```

```python
soup=BeautifulSoup(page.text,'html')
```

```python
print(soup)
```

```python
soup.find_all('th')
```

```python
soup.find_all('p')
```

```python
a=soup.find('p', class_="lead").text.strip()
```

```python
print(a)
```

```python
soup.find_all('th')
```

```python
table=soup.find_all('th')
print(table)
```

```python
table_head=[head.text.strip() for head in table]
```

```python
print(table_head)
```

```python
import pandas as pd
```

```python
df=pd.DataFrame(columns=table_head)
df
```

```python
table_row=soup.findAll("tr")
print(table_row)
```

```python
for row in table_row[1:]:
    row_data = row.find_all('td')
    # print(row_data)
    world_row_data= [data.text.strip() for data in row_data]
    # print(world_row_data)
    length= len(df)
    df.loc[length] = world_row_data
```
